##### 4.2 Arrange the complaint type

```python
df_drop.loc[:, 'Request_Closing_Hours'] = df_drop['Request_Closing_Time'].dt.total_seconds() / 3600

avg_closing_time = df_drop.groupby(['Complaint Type', 'Location Type'])['Request_Closing_Hours'].mean().reset_index()

pivot_table = avg_closing_time.pivot(index='Complaint Type', columns='Location Type', values='Request_Closing_Hours')

pivot_table.plot(kind='bar', figsize=(16, 8))
plt.title('Average Request Closing Time by Complaint Type and Location Type')
plt.xlabel('Complaint Type')
plt.ylabel('Avg Request Closing Time (Hours)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.legend(title='Location Type')
plt.show()
```

#### 5. Statistical Testing

##### 5.1. Test 1: Whether the average response time across complaint types is similar or not.

```python
from scipy.stats import f_oneway

# Drop rows with missing request closing time or complaint type
anova_data = df_drop[['Complaint Type', 'Request_Closing_Hours']].dropna()

# Group by complaint type and create a list of lists for ANOVA
grouped_data = [group['Request_Closing_Hours'].values for name, group in anova_data.groupby('Complaint Type')]

# Perform one-way ANOVA
f_stat, p_value = f_oneway(*grouped_data)

print("Test 1: ANOVA - Average Response Time Across Complaint Types")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4e}")

# Interpretation
if p_value < 0.05:
    print("Result: Reject the null hypothesis. There is a significant difference in average response time between complaint types.")
else:
    print("Result: Fail to reject the null hypothesis. No significant difference found.")
```

##### 5.2. Test 2: Whether the type of complaint or service requested and location are related.

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df_drop['Complaint Type'], df_drop['Location Type'])

# Perform Chi-square test
chi2, p_val, dof, expected = chi2_contingency(contingency_table)

print("\nTest 2: Chi-Square Test - Relationship Between Complaint Type and Location")
print(f"Chi-square statistic: {chi2:.4f}")
print(f"P-value: {p_val:.4e}")
print(f"Degrees of freedom: {dof}")

# Interpretation
if p_val < 0.05:
    print("Result: Reject the null hypothesis. Complaint type and location type are related.")
else:
    print("Result: Fail to reject the null hypothesis. No significant relationship found.")
```
